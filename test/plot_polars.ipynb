```python
import pandas as pd
import numpy as np
```

```python
from matplotlib import pyplot as plt, style
import seaborn as sns
style.use('bmh')
```

```python
def read_as_df(fname, x, y, source = None, airfoil = None):
    df = None

    if '.csv' in fname:
        df = pd.read_csv(fname)[[x, y]].copy()
    elif '.dat' in fname:
        df = pd.read_csv(fname, delim_whitespace = True)[[x, y]].copy()

    if df is None:
        raise Exception("Filename unrecognized for %s (dat or csv expected)" % (fname,))

    if source is not None:
        df['Source'] = source
    if airfoil is not None:
        df['Airfoil'] = airfoil

    return df
```

```python
def curve_contrast(
    x, y, *dfs
):
    df = pd.concat(dfs)

    sns.lineplot(
        data = df, x = x, y = y, hue = 'Airfoil', style = 'Source',
        sort = False, markers = True,
    )
```

```python
def read_airfoil(
    fname, name
):
    dat = np.loadtxt(fname, skiprows = 1)

    return {
        'name': name,
        'points': dat,
    }
```

```python
def plot_airfoils(*foils):
    for foil in foils:
        plt.plot(
            foil['points'][:, 0], foil['points'][:, 1], label = foil['name']
        )

    plt.axis('equal')

    plt.legend()

    plt.xlabel('x/c')
    plt.ylabel('y/c')
```

```python
plt.figure(figsize = (10, 4))

plot_airfoils(
    read_airfoil('23012_airfoil/n23012.dat', 'NACA-23012'), read_airfoil('2412_airfoil/n2412.dat', 'NACA-2412'), read_airfoil('4412_airfoil/n4412.dat', 'NACA-4412')
)

plt.savefig('validation_afls.pdf')
```

```python
df = pd.read_csv('time_benchmark/results.dat', delim_whitespace = True)

df = df.rename(columns = {'Precomputed(s)': 'Precomputed (s)', 'Direct(s)': 'Direct (s)', 'N': 'Number of panels'})

plt.plot(
    df['Number of panels'], df['Direct (s)'], label = 'Direct', marker = 'x'
)
plt.plot(
    df['Number of panels'], df['Precomputed (s)'], label = 'Precomputed', marker = 'x'
)

plt.yscale('log')

plt.legend()

plt.xlabel('Number of panels')
plt.ylabel('Run time (s)')

plt.savefig('runtime.pdf')
plt.show()

df['Ratio'] = df['Direct (s)'] / df['Precomputed (s)']

plt.plot(
    df['Number of panels'], df['Ratio'], marker = 'x'
)

plt.xlabel('Number of panels')
plt.ylabel('Run time ratio (direct/precomputed)')

plt.savefig('runtimeratio.pdf')
```
